### free sulfur dioxide has high standard deviation. Try to find and remove the outliers.

```python
white_wine_copy = white_wine.copy()
sns.distplot(white_wine.free_sulfur_dioxide);
sns.distplot(white_wine_copy.free_sulfur_dioxide).set_title("free_sulfur_dioxide distribution");
```

numpy.percentile to calculate 25% and 75% of free_sulfur_dioxide

```python
# calculate 25% and 75%
q25, q75 = np.percentile(white_wine_copy.free_sulfur_dioxide, 25), np.percentile(white_wine_copy.free_sulfur_dioxide, 75)
# calculate interquartile range
iqr = q75 - q25
# calculate the outlier cutoff
cut_off = iqr * 1.5
# calculate lower and upper threshold
lower, upper = q25 - cut_off, q75 + cut_off

print('iqr:' + str(iqr) + ' | cut_off:' + str(cut_off) + ' | lower:' + str(lower) + ' | upper:' + str(upper))
```

```python
# get outliers data
outliers_data_lower = [x for x in white_wine_copy.free_sulfur_dioxide if x < lower]
outliers_data_upper = [x for x in white_wine_copy.free_sulfur_dioxide if x > upper]
print("Lower: ",outliers_data_lower)
print("Upper: ",outliers_data_upper)
```

```python
white_wine_outliers_upper = white_wine_copy[white_wine_copy['free_sulfur_dioxide'].isin(outliers_data_upper)]
white_wine_outliers_upper
```

```python
# filter dataset for data lower and upper outliers
outliers_lower = white_wine_copy.free_sulfur_dioxide < lower
outliers_upper = white_wine_copy.free_sulfur_dioxide > upper
print(outliers_lower.value_counts())
print(outliers_upper.value_counts())
```

#### Plot distribution plot for both original and outliers removed datasets

```python
sns.distplot(white_wine.free_sulfur_dioxide, label='Original Dataset');
sns.distplot(white_wine_copy.free_sulfur_dioxide, label='Outliers Removed Dataset');
plt.legend();
plt.show();
```

#### correlation between free_sulfur_dioxide and quality by comparing original and outlier removed datasets

```python
correlations_original = white_wine.corr()
correlations_outliers_removed = white_wine_copy.corr()
```

```python
correlations_original["quality"]["free_sulfur_dioxide"]
```

```python
correlations_outliers_removed["quality"]["free_sulfur_dioxide"]
```

It is seen that the change in correlation is negalegible after removal of the outliers.

### total_sulfur_dioxide has high standard deviation. Try to find and remove the outliers.

##### Following the same process as earlier follow

```python
white_wine_copy = white_wine.copy()
sns.distplot(white_wine.total_sulfur_dioxide);
sns.distplot(white_wine_copy.total_sulfur_dioxide).set_title("total_sulfur_dioxide distribution");
```

```python
# calculate 25% and 75%
q25, q75 = np.percentile(white_wine_copy.total_sulfur_dioxide, 25), np.percentile(white_wine_copy.total_sulfur_dioxide, 75)
# calculate interquartile range
iqr = q75 - q25
# calculate the outlier cutoff
cut_off = iqr * 1.5
# calculate lower and upper threshold
lower, upper = q25 - cut_off, q75 + cut_off

print('iqr:' + str(iqr) + ' | cut_off:' + str(cut_off) + ' | lower:' + str(lower) + ' | upper:' + str(upper))
```

```python
# get outliers data
outliers_data_lower = [x for x in white_wine_copy.total_sulfur_dioxide if x < lower]
outliers_data_upper = [x for x in white_wine_copy.total_sulfur_dioxide if x > upper]
print("Lower: ",outliers_data_lower)
print("Upper: ",outliers_data_upper)
```

```python
# filter dataset for data upper outliers
wines_df_outliers_upper = white_wine_copy[white_wine_copy['total_sulfur_dioxide'].isin(outliers_data_upper)]
wines_df_outliers_upper
```

```python
# filter dataset for data lower outliers
wines_df_outliers_lower = white_wine_copy[white_wine_copy['total_sulfur_dioxide'].isin(outliers_data_lower)]
wines_df_outliers_lower
```

```python
# filter dataset for data lower and upper outliers
outliers_lower = white_wine_copy.total_sulfur_dioxide < lower
outliers_upper = white_wine_copy.total_sulfur_dioxide > upper
print(outliers_lower.value_counts())
print(outliers_upper.value_counts())
```

```python
# replace lower outliers with lower iqr value, upper outliers with upper iqr value 
white_wine_copy.total_sulfur_dioxide.where(~outliers_lower, lower, inplace=True)
white_wine_copy.total_sulfur_dioxide.where(~outliers_upper, upper, inplace=True)
```

```python
outliers_lower = white_wine_copy.total_sulfur_dioxide < lower
outliers_upper = white_wine_copy.total_sulfur_dioxide > upper
print(outliers_lower.value_counts())
print(outliers_upper.value_counts())
```

##### Plot distribution plot for both original and outliers removed datasets

```python
sns.distplot(white_wine.total_sulfur_dioxide, label='Original Dataset');
sns.distplot(white_wine_copy.total_sulfur_dioxide, label='Outliers Removed Dataset');
plt.legend();
plt.show();
```

##### correlation between free_sulfur_dioxide and quality by comparing original and outlier removed datasets

```python
correlations_original = white_wine.corr()
correlations_outliers_removed = white_wine_copy.corr()
```

```python
correlations_original["quality"]["total_sulfur_dioxide"]
```

```python
correlations_outliers_removed["quality"]["total_sulfur_dioxide"]
```

It is seen that the change in correlation is minor after removal of the outliers.
